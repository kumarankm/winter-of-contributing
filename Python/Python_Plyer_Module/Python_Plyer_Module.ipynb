# **Python Plyer Module**

***Plyer*** is a platform-independent api to use features commonly found on various platforms, notably mobile ones, in Python.

**What is Plyer module in python?**

Plyer module is used to access the features of the hardware. This module does not comes built-in with Python. We need to install it externally.

**Installation:**

To use on desktop: ```pip install plyer``` To use in python-for-android/kivy-ios: add ```plyer``` to your requirements if needed.

1. ***Simple Desktop Notifier***


Now with this module lets create a simple Desktop Notifier application using Python. A desktop notifier is a simple application which produces a notification message in form of a pop-up message on desktop. We will be using plyer module for the same.

In order to build notifier along with plyer, we need to import other module as well which is ```time``` module. This module works with the time object and is installed by default. You will come to know why we require it in short!:)

Here, as it is a *Desktop* notifier we have to type the below command in the terminal.

```
pip install plyer
```

Now, Import the notification class from the plyer module with following command.



```
from plyer import notification
```
This notification class comes with ```notify()``` method which consist of certain parameters.

*Syntax:*

notify(title=”, message=”, app_name=”, app_icon=”, timeout=10, ticker=”, toast=False)

*Parameters:*

* title (str) – Title of the notification
* message (str) – Message of the notification
* app_name (str) – Name of the app launching this notification
* app_icon (str) – Icon to be displayed along with the message
* timeout (int) – time to display the message for, defaults to 10
* ticker (str) – text to display on status bar as the notification arrives
* toast (bool) – simple Android message instead of full notification




```
import time
from plyer import notification
 
 
if __name__=="__main__":
 
        notification.notify(
            title = "lable to the notification",
            message=" DESCRIPTION HERE" ,
           
            # displaying time
            timeout=2
)
        # waiting time
        time.sleep(7)       # Add the sleep function to show that notification again.

```
