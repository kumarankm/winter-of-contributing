# **What is Precision Handling?**

  The precision can be defined as a numerical quantity which is a measure of the detail in which the quantity is expressed. This is usually measured in bits, but sometimes in decimal digits. It is related to precision in mathematics, which describes the number of digits that are used to express a value.

  *Handling precisions:*
There are many ways to handle/set the precision with a float value, some of the **math** functions and some of them are based on the formatting.

# **Methods to handle Precision**

How to do Precision Handling in Python?

First We will import the math module.

`import math`

Now we will see different methods.

* ***ceil():***  This method is used to return the ceiling number of the floating-point number. The ceiling number is the smallest integer greater than the number.

```
import math
number = 275.963
print('Ceiling Value: ' + str(math.ceil(number)))
```
Output

```
Ceiling Value: 276
```

* ***trunc()***: This method removes the fraction part from a floating-point number. It returns the integer part of the number.

```
import math
number = 275.963
print('Truncated value: ' + str(math.trunc(number)))
```
Output

```
Truncated value: 275
```

* ***floor():*** This method is used to return the floor number of the floating-point number. The floor number is the greatest integer smaller than the number.

```
import math
number = 275.963
print('Floor Value: ' + str(math.floor(number)))
```
Output

```
Floor Value: 275
```
Now that we have seen the code to remove the fraction part altogether. Now we will see how to manipulate the decimal part.
# **Manipulate Decimal Part (Setting Precision)**
Handing precision using different ways of formatting

* ***Using "%":***
This method is similar to “printf() function in C language”, it is used to format as well as set the precision.

```
num = 64.35498
   
# using "%" operator  
print ('The value is: %.3f'%num)     
```
Output

```
The value is: 64.355
```
* ***format():*** It is a built-in method of Python, which is used to format the string and set precision.

```
num = 64.35498354
   
# using format() function  
print ("The value is: {0:.3f}".format(num))   
```
Output

```
The value is: 64.355
```
* ***round(n,d):*** It is used to round off the number n up to d decimal places. We can select the number of digits that can be displayed after the decimal points

```
num = 64.35498354
   
# using round() function  
print ("The value is:",round(num,5))
```
Output

```
The value is: 64.35498354
```
So in conclusion, These are six basic methods to manipulate floating-point numbers.
